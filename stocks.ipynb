```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import datetime
```

```python
start_date = datetime.datetime(2018, 1, 1)
end_date = datetime.datetime(2021, 12, 31)

ford = data.DataReader("F", start=start_date, end=end_date, data_source="yahoo")
tesla = data.DataReader("TSLA", start=start_date, end=end_date, data_source="yahoo")
gm = data.DataReader("GM", start=start_date, end=end_date, data_source="yahoo")
```

```python
# There's a problem here!
plt.plot(ford.index, ford["Adj Close"], label="Ford", color="blue")
plt.plot(tesla.index, tesla["Adj Close"], label="Tesla", color="orange")
plt.plot(gm.index, gm["Adj Close"], label="GM", color="green")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Let's fix the problem
fig, ax1 = plt.subplots()

fig.suptitle("Stock Prices")

ax1.set_xlabel("Date")
ax1.set_ylabel("Price")
ax1.plot(ford.index, ford["Adj Close"], label="Ford", color="blue")
ax1.plot(gm.index, gm["Adj Close"], label="GM", color="green")

ax1.tick_params(axis="x", rotation=45)

# instantiate a second axis that shares the same as x-axis
ax2 = ax1.twinx()

color = "orange"

ax2.set_ylabel("Tesla Price", color=color)

ax2.plot(tesla.index, tesla["Adj Close"], label="Tesla", color=color)

# ax2.tick_params(axis="y", labelcolor=color)
ax2.tick_params(axis="x", rotation=45)

plt.tight_layout()
plt.show()
```
