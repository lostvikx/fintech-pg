```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("A single chart")
```

```python
# .subplots(nRows, nCols)
fig, axs = plt.subplots(2)
fig.suptitle("Vertically stacked subplots")
axs[0].plot(x, y)
axs[1].plot(x, -y) # Just the inverse of y axis
```

```python
# An alternate method
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle("Vertically stacked subplots")
ax1.plot(x, y)
ax2.plot(x, -y)
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y, "orange")
axs[0, 0].set_title("Axis [0, 0]")
axs[0, 1].plot(x, -y, "blue")
axs[0, 1].set_title("Axis [0, 1]")
axs[1, 0].plot(-x, y, "green")
axs[1, 0].set_title("Axis [1, 0]")
axs[1, 1].plot(-x, -y, "red")
axs[1, 1].set_title("Axis [1, 1]")

for ax in axs.flat:
  ax.set(xlabel="x-label", ylabel="y-label")
  ax.label_outer()

plt.tight_layout()
```

```python
fig, (ax1, ax2) = plt.subplots(2, sharex=True)
fig.suptitle("Aligned x-axis sharex")
ax1.plot(x, y)
ax2.plot(x + 2, -y)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
fig.suptitle("Aligned y-axis sharey")
ax1.plot(x, y + 2)
ax2.plot(x, -y)
```
