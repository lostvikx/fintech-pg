```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
student_heights = [150, 187, 168, 175, 164]
student_names = ["Jake", "Vik", "Kate", "Josh", "Ali"]

plt.plot(np.arange(len(student_names)), student_heights)
# plt.bar(np.arange(len(student_names)), student_heights)

plt.yticks(np.arange(100, 201, 20))

plt.show()
```

```python
# Stacked bar
men_mean = [20, 25, 40, 35, 33]
women_mean = [25, 32, 18, 20, 22]

# use index numbers instead of stings on x axis
index = np.arange(len(men_mean))
width = 0.5

p1 = plt.bar(index, men_mean, width)
p2 = plt.bar(index, women_mean, width, bottom=men_mean)

plt.ylabel("scores")
plt.xlabel("x data")

plt.title("Scores by group & gender")
plt.xticks(index, ["Gr 1", "Gr 2", "Gr 3", "Gr 4", "Gr 5"])
plt.yticks(np.arange(0, 81, 10))

# we can just use the just the 0th index instead of the entire the bar data structure
plt.legend((p1[0], p2[0]), ("men", "women"))

plt.show()
```

```python
# Bar joint
labels = ["Gr 1", "Gr 2", "Gr 3", "Gr 4", "Gr 5"]
men_mean = [20, 25, 40, 35, 33]
women_mean = [25, 32, 18, 20, 22]

# use index numbers instead of stings on x axis
index = np.arange(len(labels))
width = 0.25

# axes
fig, ax = plt.subplots()
rect1 = ax.bar(index - width/2, men_mean, width, label="Men")
rect2 = ax.bar(index + width/2, women_mean, width, label="Women")

ax.set_ylabel("Score")
ax.set_xlabel("Groups")
# important to set the indeices as the xticks, then set the lables
ax.set_xticks(index)
ax.set_xticklabels(labels)
ax.set_title("Scores by group & gender")
ax.legend()

plt.show()
```

```python
# Categorical Plotting
group_names = ["noclap", "anizone", "vixens"]
values = [1, 10, 100]

# plt.bar(group_names, values)

plt.figure(figsize=(9, 3))

# the nRows, nCols, index
plt.subplot(131)
plt.bar(group_names, values)
plt.title("Bar chart")

plt.subplot(132)
plt.plot(group_names, values, "--x")
plt.title("Line chart")

plt.subplot(133)
plt.scatter(group_names, values)
plt.title("Point chart")

plt.suptitle("Categorical Plotting")
plt.tight_layout()

plt.show()
```

```python
# Histogram
mu, sigma = 100, 15
# creates 10000 length array with random data, near the mean
x = mu + sigma * np.random.randn(10000)

# bins are the range groups
n, bins, patches = plt.hist(x, 50, density=0, facecolor="g", alpha=0.75)

plt.xlabel("IQ")
plt.ylabel("Frequency")
plt.title("Histogram for IQ")

# text inside the canvas
plt.text(60, 500, r"$\mu=100,\ \sigma=15$")

# (x_start, x_end, y_start, y_end)
plt.axis([40, 160, 0, 700])

plt.grid(False)

plt.show()
```

```python
# Pie chart
labels = ["Dogs", "Cats", "Birds", "Monkeys"]
sizes = [35, 30, 15, 20]
print(sum(sizes) == 100)

explode = (0, 0, 0.1, 0)
fig1, ax1 = plt.subplots()

# actopct is the percentage over the slices
ax1.pie(sizes, explode, labels, autopct="%.2f%%", shadow=False, startangle=45)
ax1.axis("equal")

plt.show()
```
